```python
age_counter = data.Age.value_counts().head(7)

age_counter
```

```python
bin_edges = [0,1,2,3]
plt.hist(age_counter,
         bins= bin_edges,
         density=False,
         histtype='bar',
         color='b',
         edgecolor='k',
         alpha=0.5)

plt.xlabel('age counter')
plt.xticks([0,1,2,3])
plt.ylabel('Number of interviewee')
plt.title('Histograme of age counter')

plt.show()
```

```python
gender_distinct= data.Gender.value_counts().head(10)
gender_distinct
```

```python
plt.figure(figsize=(12,6))
plt.title('Gender')
sns.barplot(x=gender_distinct.index, y=gender_distinct);
```

```python
plt.figure(figsize=(20,20))
plt.plot(data['Product rate'], data['Importance'], 's-b')

plt.xlabel('Product rate')
plt.ylabel('Importance')

plt.title('Importance and Rate');
```
